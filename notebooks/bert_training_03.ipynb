```python
import torch

print("Torch version:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())
```

```python
from datasets import Dataset
from transformers import (
    DistilBertTokenizerFast,
    DistilBertForSequenceClassification,
    Trainer,
    TrainingArguments,
    DataCollatorWithPadding
)

print("✅ Transformers import successful")
```

```python

import pandas as pd
import numpy as np

from sklearn.metrics import accuracy_score, f1_score, classification_report,confusion_matrix
```

**Load Your Train & Validation Data**

```python
train_df = pd.read_csv("../data/processed/train_baseline.csv")
val_df   = pd.read_csv("../data/processed/val_baseline.csv")

print(train_df.shape, val_df.shape)
train_df.head()
```

**Basic Safety Checks**

```python
# Drop missing values (just in case)
train_df = train_df.dropna(subset=["text", "label"])
val_df   = val_df.dropna(subset=["text", "label"])

# Ensure correct types
train_df["label"] = train_df["label"].astype(int)
val_df["label"]   = val_df["label"].astype(int)

print("Labels in train:", sorted(train_df["label"].unique()))
print("Labels in val:", sorted(val_df["label"].unique()))
```

**Convert Pandas → HuggingFace Dataset**

```python
train_dataset = Dataset.from_pandas(
    train_df[["text", "label"]],
    preserve_index=False
)

val_dataset = Dataset.from_pandas(
    val_df[["text", "label"]],
    preserve_index=False
)
```

**Load Tokenizer**

```python
tokenizer = DistilBertTokenizerFast.from_pretrained(
    "distilbert-base-uncased"
)
```

**Tokenization**

```python
def tokenize_function(batch):
  return tokenizer(
      batch['text'],
      truncation=True,
      padding=False,
      max_length=128
  )

train_dataset=train_dataset.map(
    tokenize_function,
    batched=True,
    remove_columns=["text"]
)

val_dataset=val_dataset.map(
    tokenize_function,
    batched=True,
    remove_columns=["text"]
)
```

**Data Collator (Dynamic Padding)**

```python
data_collator=DataCollatorWithPadding(tokenizer=tokenizer)
```

**Load DistilBERT Model**

```python
num_labels=train_df['label'].nunique()
print("Number of labels:", num_labels)
```

```python
model=DistilBertForSequenceClassification.from_pretrained(
    'distilbert-base-uncased',
    num_labels=num_labels
)
```

**Metrics Function (For Evaluation)**

```python
def compute_metrics(eval_pred):
  logits, labels= eval_pred
  preds= np.argmax(logits, axis=1)

  return {
      "accuracy": accuracy_score(labels, preds),
      "f1_macro":f1_score(labels,preds, average='macro')
  }
```

**Training Arguments**

```python
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,   # reduce to 8 if CPU is slow
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=200,
    load_best_model_at_end=True,
    metric_for_best_model="f1_macro",
    report_to="none"
)
```

```python
import accelerate
print(accelerate.__version__)
```

**Trainer Setup**

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)
```

**TRAIN THE MODEL**

```python
trainer.train()
```

**Final Evaluation Report**

```python
# Run prediction
preds_output = trainer.predict(val_dataset)

# True labels
y_true = preds_output.label_ids

# Predicted labels
y_pred = np.argmax(preds_output.predictions, axis=1)

print(classification_report(y_true,y_pred))
```

```python
cm = confusion_matrix(y_true, y_pred)
print(cm)
```

**Save Model**

```python
trainer.save_model("./mental_health_distilbert")
tokenizer.save_pretrained("./mental_health_distilbert")
```
